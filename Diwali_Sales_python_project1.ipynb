```python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline
sns.set(color_codes=True)
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
# change data type
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.describe()
```

```python
df[['Age', 'Orders', 'Amount']].describe()
```

```python
# plotting a bar chart for Gender and it's count
ax = sns.countplot(x = 'Gender',data = df)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

```python
##OBSERVATION :- Now we can see that most of the buyers are females and even the purchasing power of females are greater than men.
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

```python
##OBSERVATION :- now we can see that most of the buyer are in age group 26-35 yrs females
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(5)

sns.barplot(x = 'State',y= 'Orders' ,data = sales_state)
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5)

sns.barplot(x = 'State',y= 'Amount' ,data = sales_state)
```

```python
##OBSERVATION :- most of the orders and total sales amount are from Uttar Pradesh, Maharashtra and Karnataka respectively.
```

```python
ax = sns.countplot(data=df, x='Marital_Status', hue = 'Gender')
```

```python
sales_marital_status = df.groupby(['Marital_Status', "Gender"], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Marital_Status',y= 'Amount', hue='Gender',data = sales_marital_status)
```

```python
##OBSERVATION :- Most of the buyers are married females and they have high purchased amount.
```

```python

ax = sns.countplot(data=df, x='Occupation')
sns.set(rc={'figure.figsize':(10,5)})
```

```python
sales_occupation = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Occupation',y= 'Amount',data = sales_occupation)
```

```python
##OBSERVATION :- Most of the buyers are working in IT, Healthcare and aviation
```

```python
ax = sns.countplot(data=df, x='Product_Category')
sns.set(rc={'figure.figsize':(20,10)})
```

```python
sales_Product_Category = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5)

sns.barplot(x = 'Product_Category',y= 'Amount',data = sales_Product_Category)
```

```python
##OBSERVATION :- most of the sold product are from food, clothing and electronics.
```

```python
sales_product = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(5)

sns.barplot(x = 'Product_ID',y= 'Orders' ,data = sales_product)
```

```python
##CONCLUSION:- Married women age group 26-35 yrs from UP, Maharashtra and Karnataka working in IT, Healthcare and Aviation are most likely buy product from foods, clothing and electronics category.
```
